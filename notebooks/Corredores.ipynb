```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/OscarPasasin/etl-data-pipeline/refs/heads/main/data/raw/corredores.csv"

df = pd.read_csv(url)

df.head()
```

```python
#Exploracion de datos
df.shape
df.columns
df.info()
df.isnull().sum()
```

```python
#Limpieza de datos
corredores=df.copy()

print(corredores['zona'].value_counts(dropna=False))
```

```python
corredores['zona'] = corredores['zona'].str.strip().str.title()
print(corredores['zona'].value_counts(dropna=False))
```

```python
corredores['nivel'] = corredores['nivel'].str.strip().str.lower()
print(corredores['nivel'].value_counts(dropna=False))
```

```python
corredores['anios_experiencia']=pd.to_numeric(corredores['anios_experiencia'],errors='coerce')
print(corredores['anios_experiencia'].value_counts(dropna=False))
```

```python
#Separar datos válidos y rechazados
filtrado_validos = (corredores['anios_experiencia'].notna()) & \
(corredores['nivel'].notna()) & \
(corredores['zona'].notna())
validos = corredores[filtrado_validos].copy()
rechazados = corredores[~filtrado_validos].copy()
```

```python
#Motivo de rechazo
def motivo_rechazo(row):
  motivos = []
  if pd.isnull(row['anios_experiencia']):
    motivos.append('anios_experiencia_vacio')
  if pd.isnull(row['nivel']):
    motivos.append('nivel_vacio')
  if pd.isnull(row['zona']):
    motivos.append('zona_vacio')
  return ','.join(motivos)

rechazados['motivo_rechazo'] = rechazados.apply(motivo_rechazo, axis=1)
```

```python
validos.to_csv("corredores_curated.csv", index=False)

rechazados.to_csv("corredores_rejects.csv", index=False)
```
