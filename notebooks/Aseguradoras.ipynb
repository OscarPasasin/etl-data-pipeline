```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/OscarPasasin/etl-data-pipeline/refs/heads/main/data/raw/aseguradoras.csv"

df = pd.read_csv(url)

df.head()
```

```python
#Exploracion de datos
df.shape
df.columns
df.info()
df.isnull().sum()
```

```python
#Limpieza de datos
aseguradoras=df.copy()
print(aseguradoras['pais'].value_counts(dropna=False))
```

```python
aseguradoras['pais'] = aseguradoras['pais'].replace('ElSalvador', 'El Salvador')
print(aseguradoras['pais'].value_counts(dropna=False))
```

```python
asegudoras=df.copy()
print(aseguradoras['rating_riesgo'].value_counts(dropna=False))
```

```python
#Separar datos válidos y rechazados
validos = aseguradoras[
    aseguradoras['pais'].notna() &
    aseguradoras['rating_riesgo'].notna()
].copy()

rechazados = asegudoras[asegudoras['pais'].isna() |
aseguradoras['rating_riesgo'].isna()
].copy()
```

```python
#Motivo de rechazo
def motivo(row):
    motivos = []
    if pd.isna(row['pais']):
        motivos.append("pais_vacio")
    if pd.isna(row['rating_riesgo']):
        motivos.append("rating_riesgo_vacio")
    return ",".join(motivos)

rechazados["motivo_rechazo"] = rechazados.apply(motivo, axis=1)
```

```python
validos.to_csv("aseguradoras_curated.csv", index=False)

rechazados.to_csv("aseguradoras_rejects.csv", index=False)
```
